# Why Polynomial Regression?

- To handle non-linear data

# How Does Polynomial Regression Handle Non-Linear Data?

- The power is known as the Degree and is considered a Hyper-parameter
- A higher value Degree results in overfitting, a lower values becomes less fitted
- Polynomial Regression models are usually fitted with the method of "least squares"
- The "least square" method minimizes the "variance" error

# Why Is Polynomial Regression Called Polynomial Linear Regression?

- So we're actually trying to estimate the coefficients and y.
- X and y are given to us, so we're working on figuring out the coefficients, or the Degrees (aka. power)
- The degree of 1 represents simple linear regression.

# Linear Regression Vs. Polynomial Regression

## Polynomial Regression With One Variable

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
X = 6 * np.random.rand(200, 1) - 3
y = 0.8 * X**2 + 0.9*X + 2 + np.random.randn(200, 1)
```

```python
X[:5]
```

```python
#equation used -> y = 0.8x^2 + 0.9x + 2
#visualize the data
plt.plot(X, y, 'b.')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

# Apply simple linear regression

```python
lr = LinearRegression().fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

```python
lr.score(x_test, y_test)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.plot(x_train, lr.predict(x_train), color="r")
plt.plot(X, y, "b.")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

# Apply polynomial regression

```python
poly = PolynomialFeatures(degree=2, include_bias=True)
```

```python
x_train_trans = poly.fit_transform(x_train)
x_test_trans = poly.transform(x_test)
```

```python
poly_lr = LinearRegression().fit(x_train_trans, y_train)
```

```python
y_pred = poly_lr.predict(x_test_trans)
```

```python
r2_score(y_test, y_pred)
```

```python
poly_lr.coef_
```

```python
poly_lr.intercept_
```

```python
X_new = np.linspace(-3, 3, 200).reshape(200, 1)
X_new_poly = poly.transform(X_new)
y_new = poly_lr.predict(X_new_poly)
```

```python
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.plot(x_train, y_train, "b.",label='Training points')
plt.plot(x_test, y_test, "g.",label='Testing points')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
