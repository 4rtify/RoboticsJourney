### What is an Equation?
- An equation says two things are equal. It will have an equals sign.
- What is on the left is equal to what is on the right.

```python
import sympy as sp
x = sp.Symbol('x')
equation = sp.Eq(3*x + 9, 45)
display(equation)
display(sp.solve(equation, x))
```

### Subject of an equation

```python
import sympy as sp
x, y, a, b = sp.symbols('x y a b')
equation = sp.Eq(y, a*x + b)
display(equation)
# Solve for x
x_expression = sp.solve(equation, x)
display(x_expression)
```

### What is a formula?
- A formula is a fact or rule that uses mathematical symbols.
- It will usually have an equals sign and two or more variables that stand in for values we don't know yet.
- It shows us how things are related to each other.
- Example of the formula for finding the volume of a box.

```python
import sympy as sp
length, width, height = sp.symbols('length width height')
volume = length * width * height
display(volume.subs({length: 10, width: 4, height: 5}))
```

### Subject of a formula
- Changing the subject of the formula from Volume to Width 

```python
import sympy as sp
l, w, h, V = sp.symbols('l w h V')
volume_formula = sp.Eq(V, l * w * h)
width_formula = sp.solve(volume_formula, w)
display(width_formula)
```

### Commutative, Associative and Distributive Laws
1. The Commutative Laws say we can swap numbers over and still get the same result   
a + b = b + a  
a*b = b*a
- Why "commutative"? Because the numbers can travel back and forth like a commuter

2. Associative Laws
- The law says it doesn't matter how we group the numbers (which we calculate first)  
(a + b) + c = a + (b + c)  
(a  b)  c  =  a  (b  c)
