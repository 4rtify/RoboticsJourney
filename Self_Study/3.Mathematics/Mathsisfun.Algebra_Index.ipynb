### What is an Equation?
- An equation says two things are equal. It will have an equals sign.
- What is on the left is equal to what is on the right.

```python
import sympy as sp
x = sp.Symbol('x')
equation = sp.Eq(3*x + 9, 45)
display(equation)
display(sp.solve(equation, x))
```

### Subject of an equation

```python
import sympy as sp
x, y, a, b = sp.symbols('x y a b')
equation = sp.Eq(y, a*x + b)
display(equation)
# Solve for x
x_expression = sp.solve(equation, x)
display(x_expression)
```

### What is a formula?
- A formula is a fact or rule that uses mathematical symbols.
- It will usually have an equals sign and two or more variables that stand in for values we don't know yet.
- It shows us how things are related to each other.
- Example of the formula for finding the volume of a box.

```python
import sympy as sp
length, width, height = sp.symbols('length width height')
volume = length * width * height
display(volume.subs({length: 10, width: 4, height: 5}))
```

### Subject of a formula
- Changing the subject of the formula from Volume to Width 

```python
import sympy as sp
l, w, h, V = sp.symbols('l w h V')
volume_formula = sp.Eq(V, l * w * h)
width_formula = sp.solve(volume_formula, w)
display(width_formula)
```

## Equation of a straight line

- The equation for a straight line is usually written this way
  - y = mx + b
  - y is how far up
  - x is how far along
  - m is the Slope or Gradient (how steep the line is)
  - b is the value of y when x = 0

- How do we find "m" and "b"?
  - b "Y Intercept": Where the line crosses the Y axis when X is zero
  - m "Slope": Change in Y / change in X





- Part 1: So from the above pictures:
  - b = 1 (value of Y when X = 0) - AKA the Y Intercept I'm pretty sure
  - m = 2 / 1 = 2
- y = mx + b
  - y = 2x + 1

- Part 2: With the above equation, we can choose any value for X and find the matching value for Y
- When x = 1
  - y = 2 * 1 + 1
  - y = 3
- When x = 7
  - y = 2 * 7 + 1
  - y = 15

## Vertical Line
- What is the equation for a vertical line?
- The slope is undefined, and where does it cross the Y axis?

## Rise and Run
- Sometimes the words rise and run are used
  - Rise is how far up
  - Run is how far along
- So the slope is:
  - m = rise / run
  - m = change in Y / change in X
