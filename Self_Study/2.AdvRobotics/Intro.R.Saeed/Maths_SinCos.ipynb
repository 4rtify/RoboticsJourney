```python
import numpy as np
import math
```

```python
theta = math.radians(45) 

R = np.array([
    [math.cos(theta), -math.sin(theta)],
    [math.sin(theta), math.cos(theta)]
])
R
```

```python
point = np.array([1, 0])

rotated_point = np.dot(R, point)

print("Original point:", point)
print("Rotated point:", rotated_point)
```

# Ratio

# Sine, Cosine and Tangent

- For angle 'theta', the functions are calculated this way:
- Sine Function: sin(theta) = Opposite / Hypotenuse
- Cosine Function: cos(theta) = Adjacent / Hypotenuse
- Tangent Function: tan(theta) = Opposite / Adjacent

# Unit Circle

- For any angle 'theta', the coordinates of the point on the unit circle are (cos(theta), sin(theta))
- When you input an angle in radians into the sine function, it essentially finds the Y coordinate of the point on the unit circle.

Why lengths are not needed:
- The sine function inherently knows it's dealing with a unit circle with a radius (hypotenuse) of 1

# Visual Understanding

- Imagine a right angle triangle inscribed into a unit circle
- It has an angle of 35 degrees and the hypotenuse has a length of 1 unit
- The "opposite" side is the vertical line from the X axis to the point on the circle corresponding to 35 degrees
- The length of this "opposite" side is calculated by sin(35)
- Which is 0.57 units
- So by default the sin function uses the Unit Circle with a radius of 1 to calculate an answer

```python
math.sin(math.radians(35))
```
