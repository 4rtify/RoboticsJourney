# 2. Cross Product (result is a vector) # Vector Product

- A cross product is another vector
- The cross product a x b of two vectors is another vector that is at right angles to both



- The magnitude (length) of the cross product equals the area of a parallelogram with vectors a and b:



- The cross product (in blue) is:
1. Zero in length when vectors A and B point in the same or opposite direction
2. Reaches maximum length when vectors A and B are at right angles




Calculating:

a x b = |a| * |b| * sin(theta) * n

- |a| is the magnitude of vector a
- |b| is the magnitude of vector b
- theta is the angle between a and b
- n is the unit vector at right angles to both a and b





```python
import numpy as np
import math
a = np.array([2,3,4])
b = np.array([5,6,7])

cx = (a[1] * b[2]) - (a[2] * b[1])
cy = (a[2] * b[0]) - (a[0] * b[2])
cz = (a[0] * b[1]) - (a[1] * b[0])

c = np.array([cx, cy, cz])
c
```

# Converting between Cartesian and Polar Coordinates

- If we know the cartesian coordinate, and want to convert to Polar we use a right angle triangle.
- We use Pythagoreans Theorem 
- r<sup>2</sup> = x<sup>2</sup> + y<sup>2</sup>

```python
import sympy as sp
r, x, y = sp.symbols('r x y')
equation = sp.Eq(r**2, x**2 + y**2)
display(equation)
```

Use Pythagoras to find the Hypotenuse

```python
import sympy as sp
r = sp.Symbol('r')
x = 12
y = 5

equation = sp.Eq(r**2, x**2 + y**2)
r_solution = sp.solve(equation, r)
display(r_solution)
```

Use Tangent Function to find the angle

```python
theta_radians = math.atan(y / x)
```

```python
theta_degrees = math.degrees(theta_radians)
```

```python
round(theta_degrees,1)
```

# Converting from Polar to Cartesian

- We solve a right angle triangle with a known long side and angle

```python
print("Solving for X:")
print(f"{theta_degrees}")
print(f"{r_solution}")

r_solution[1] * math.cos(math.radians(theta_degrees))
```

```python
print("Solving for Y:")
r_solution[1] * math.sin(math.radians(theta_degrees))
```
