## 2.1.1 Relative pose

The combination of position and orientation = pose  
The blue object moves position to become the red object, this is known as being transformed - the blue object transformed into the pose of the red object  

- It is natural to think of this as motion  
- Also the orientations has changed, so it has rotated.
- Represented as the symbol ksigh ξ - pronouced z-eye
- The leading super script indicates the initial pose  
- The trailing sub script indicates the resulting pose  
- Therefore $^xξ_y$ means the motion from pose x to pose y

Motions can be joined up to create arbitary motions - this is known as compsition or compounding. And written algebraically as:  

This is the circled plus symbol: $ \oplus $  

$^0ξ_2$ = $^0ξ_1$   $\oplus$   $^1ξ_2$

So the 2D figures show the translations.  

The 3D example also shows the rotation which is a richer set of motions.  

The order of motions is important - example of turning 90 degree and walking 1m is different to walking 1m first, then turning 90 degree.  

The below motions and inverse or opposite motions gets us back to where we started.  
$^xξ_y$ $\oplus$ $^yξ_x$ = $^xξ_x$ = $\varnothing$  

Null motion means $\varnothing$

$a \equiv b$  
Note: the equal sign is used for specific instances of equality, while the identically equal sign denotes an identity or a general truth that holds universally  

The inverse symbol is $\ominus$  

$^xξ_y$ $\equiv$ $\ominus$ $^yξ_x$  

There is no absolute way to describe the pose of an object, we can only describe its pose with respect to some other pose, and we introduce a reference pose for this purpose. Any pose is always a relative pose, described by the motion required to get there from the reference pose.

**<u>Basis Vectors</u>**  
- Definition: Basis vectors are a set of vectors that define a coordinate system. They serve as reference directions for that system.  
- Purpose: They are used to describe the position of any point in that space by specifying how much you need to move along each of these reference directions.  
- Example in 2D: In a 2D coordinate system, the basis vectors are usually:  
- 𝑥^: Points in the direction of the x-axis.  
- 𝑦^: Points in the direction of the y-axis.  

**<u>Unit Vectors</u>**   
- Definition: Unit vectors are vectors with a length (or magnitude) of 1.  
- Purpose: They are used to indicate direction without specifying length. Any vector can be expressed as a combination of unit vectors scaled by the appropriate amounts.  
- Example: In a 2D coordinate system:  
- x^: is a unit vector pointing along the x-axis.  
- y^: is a unit vector pointing along the y-axis.

### 2.2.1.2 Matrix Exponential for Rotation

There is a fascinating, and very useful, connection between a rotation matrix and the "exponential of a skew-symmetric matrix".  
We can easily demonstrate this by considering, again, a pure rotation of 0.3 radians expressed as a rotation matrix.

```python
from scipy.linalg import logm 
import numpy as np

R = rot2(0.3)
print(f"1. R for rotational matrix is: \n{R}\n")

L = logm(R)  
print(f"2. L for logarithm of R is: \n{L}")
```

```python
print("1. This matrix is an example of a 2 x 2 skew-symmetric matrix. It has only one unique element")

S = vex(L)

print(f"S is: \n\t{S}")

print("2. which is the rotation angle as a single-element array, not a scalar. \nThis is our first encounter with Lie (pronounced lee) group theory, and we will continue to explore this topic in the rest of this chapter.")
```
